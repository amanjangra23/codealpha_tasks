```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Reading imported CSV file**

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="latin1")
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.drop(["unnamed1","Status"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
grouped_data=df.groupby("Gender")["Orders"].value_counts().unstack()
print(grouped_data)
```

```python
grouped_data=df.groupby("Gender")["Orders"].value_counts().unstack()
grouped_data.plot(kind="bar")
plt.title("distribution of orders placed by gender")
plt.show()
```

```python
grouped_data=df.groupby("Marital_Status")["Orders"].value_counts().unstack()
grouped_data.plot(kind="bar")
plt.show()
```

```python
grouped_data=df.groupby("State")["Orders"].value_counts().unstack()
grouped_data.plot(kind="bar",figsize=(20,10))
plt.show()
```
